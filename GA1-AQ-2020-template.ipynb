# Introduction

Reports such as [this one](https://www.nature.com/articles/s41893-020-0581-y) show that the COVID-19 lockdown in China at least had one benefit: suddenly the air quality in the country had drastically improved. Possible explanations are the great reduction in traffic and the large-scale closing of industries. Also in the Netherlands, life has drastically changed, with for instance many more people working from home. This drastic situation change between before and after COVID-19, may also allow us to get a better understanding as to what factors do (or do not) influence the air quality in the Netherlands, especially if we combine the efforts of all DAE students in 2020-2021.

We will use air quality data from [**luchtmeetnet.nl**](https://luchtmeetnet.nl). This website publishes the concentrations of various chemical compounds from approximately one hundred measurement stations in the Netherlands. With this data we will aim to quantify differences in air quality before and after the onset of COVID-19.

But even after quantifying the difference in air quality, how could we say the difference is large or small? For this we will compare it with yearly, seasonal and daily variations in air quality.

Finally, we will take into account several other factors that may influence the air quality, such as the amount of traffic and the weather. The traffic data is an aggregation of data downloaded from the [**ndw**](https://www.ndw.nu/en/), the national data warehouse on traffic information. This website stores historical traffic data measured at thousands of different locations in the Netherlands. The weather data is retrieved from the KNMI.
